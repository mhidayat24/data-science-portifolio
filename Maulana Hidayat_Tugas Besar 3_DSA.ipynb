# Soal 1: Baca data vgsales

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go




#code di sini
vgs = pd.read_csv('/content/vgsales.csv')
vgs.head()
```

Expected Output:
soal 1 tugas 3.png

<h2>Soal 2: Preparing Data</h2>

Manipulasikanlah data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code di sini
vgs.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
```

Expected Output:

soal 2 tugas 3.png

Soal 3: Tranformasikan data tersebut,dengan harapan saat divisualisasi  NA_Sales,EU_Sales,JP_Sales dan Other_Sales ada di sumbu-x

```python
#code di sini


vgs2 = vgs.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean().T.reset_index()
vgs2 = vgs2.rename({'index':'Genre'},axis=1)
vgs2 = vgs2.rename_axis(None,axis=1)
vgs2
```

Expected Output:

soal 3 tugas 3.png

---

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>
1. Visualisasikan gambar yang tadi kamu manipulasi, untuk mengjasilkan visualisasi seperti di bawah ini
2. Tuliskan kesimpulan yang didapat dari visualisasi tersebut

```python
#code di sini

fig = go.Figure(data=[
                      go.Bar(name='Action',x=vgs2.Genre,y=vgs2['Action']),
                      go.Bar(name='Adventure',x=vgs2.Genre,y=vgs2['Adventure']),
                      go.Bar(name='Fighting',x=vgs2.Genre,y=vgs2['Fighting']),
                      go.Bar(name='Misc',x=vgs2.Genre,y=vgs2['Misc']),
                      go.Bar(name='Platform',x=vgs2.Genre,y=vgs2['Platform']),
                      go.Bar(name='Puzzle',x=vgs2.Genre,y=vgs2['Puzzle']),
                      go.Bar(name='Racing',x=vgs2.Genre,y=vgs2['Racing']),
                      go.Bar(name='Role-Playing',x=vgs2.Genre,y=vgs2['Role-Playing']),
                      go.Bar(name='Shooter',x=vgs2.Genre,y=vgs2['Shooter']),
                      go.Bar(name='Simulation',x=vgs2.Genre,y=vgs2['Simulation']),
                      go.Bar(name='Sports',x=vgs2.Genre,y=vgs2['Sports']),
                      go.Bar(name='Strategy',x=vgs2.Genre,y=vgs2['Strategy'])
                      
])
fig.update_layout(
    
    title = 'Mean Sales Video Games by Genre',
    xaxis_title='Regional Sales',
    yaxis_title='Mean Sales')

fig.show()
```

```python

ax=vgs2.plot(x='Genre',kind='bar',stacked=False,rot=0)
fig = ax.get_figure()
fig.set_size_inches(20,10)
plt.xlabel('Regional Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games by Genre')
plt.show()
```

Expected Output(abaikan warna pada grafik):
soal 4 tugas 3.png


Kesimpulan:

Mean tertinggi di regional NA bergenre Platform, mean tertinggi di regional EU bergenre Shooter, mean tertinggi di regional JP bergenre Role Playing, mean tertinggi di regional Other bergenre Shooter

<h2>Soal 5: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kamu manipulasi, untuk mengjasilkan visualisasi seperti di bawah ini
2. Tuliskan kesimpulan yang didapat dari visualisasi tersebut

```python
#code di sini

fig = go.Figure(data=[
                      go.Bar(name='Action',x=vgs2.Genre,y=vgs2['Action']),
                      go.Bar(name='Adventure',x=vgs2.Genre,y=vgs2['Adventure']),
                      go.Bar(name='Fighting',x=vgs2.Genre,y=vgs2['Fighting']),
                      go.Bar(name='Misc',x=vgs2.Genre,y=vgs2['Misc']),
                      go.Bar(name='Platform',x=vgs2.Genre,y=vgs2['Platform']),
                      go.Bar(name='Puzzle',x=vgs2.Genre,y=vgs2['Puzzle']),
                      go.Bar(name='Racing',x=vgs2.Genre,y=vgs2['Racing']),
                      go.Bar(name='Role-Playing',x=vgs2.Genre,y=vgs2['Role-Playing']),
                      go.Bar(name='Shooter',x=vgs2.Genre,y=vgs2['Shooter']),
                      go.Bar(name='Simulation',x=vgs2.Genre,y=vgs2['Simulation']),
                      go.Bar(name='Sports',x=vgs2.Genre,y=vgs2['Sports']),
                      go.Bar(name='Strategy',x=vgs2.Genre,y=vgs2['Strategy']),
                      
])
fig.update_layout(
    barmode='stack',
    title = 'Mean Sales Video Games by Genre',
    xaxis_title='Regional Sales',
    yaxis_title='Mean Sales')
fig.show()
```

```python
plt.figure(figsize=(20,10))
ax=vgs2.plot(x='Genre',kind='bar',stacked=True,rot=0)
fig = ax.get_figure()
fig.set_size_inches(20,10)
plt.xlabel('Regional Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games by Genre')
plt.show()
```
